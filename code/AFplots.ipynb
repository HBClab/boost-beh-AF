```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

#### Read in csv

```python
df = pd.read_csv('/Users/zakg04/Documents/HBC_lab/BOOST/Main/test-7005/AF/7005_AF_A.csv')
df
```

```python
# separate practice from test data using second column
test = df[df['block']=='test']

test
```

```python
# calculate percent responded over all trials by dividing number of columns with a letter in 'response' by number of trials

percent_responded = test['response'].apply(lambda x: len(x)/len(test['trial']))
lambda_correction = 1 - percent_responded.mean()
if lambda_correction > 0.99:
    lambda_correction = 1
lambda_correction
```

```python
#calculate percent correct by dividing number of 1s in correct column by number of trials
percent_correct = test['correct'].sum()/len(test['trial'])
percent_correct
```

```python
#calculate mean RT
mean_rt = test['response_time'].mean()
mean_rt
```

```python
# create a scatter plot of response time by condition, use transparency to show density, add transparent box and whiskers
plt.figure(figsize=(10,6))
sb.stripplot(x='condition', y='response_time', data=test, alpha=0.5)
sb.boxplot(x='condition', y='response_time', data=test, whis=np.inf, linewidth=0.5)
plt.show()
```

```python
#create a bar chart showing only the percentage of trials completed, red is less than 50% completed, orange is less than 85% completed, green is 85% or more completed

if len(test['trial'])/len(df['trial']) < 0.5:
    color = 'red'
elif len(test['trial'])/len(df['trial']) < 0.85:
    color = 'orange'
else:
    color = 'green'

PC = sb.barplot(x=['Percent Answered', 'Percent Correct'], y=[lambda_correction, percent_correct], color=color, width=0.2, native_scale=False)
PC.bar_label(PC.containers[0], label_type='edge')
plt.show()
```

```python

# Plotting the circular bar graph
def plot_circular_bar_graph(percentages, name):
    startangle = 90
    colors = ['#4393E5', '#43BAE5', '#7AE6EA', '#E5A443']
    
    # Convert data to fit the polar axis
    ys = [i *1.1 for i in range(len(percentages))]   # One bar for each block
    left = (startangle * pi * 2) / 360  # This is to control where the bar starts

    # Figure and polar axis
    fig, ax = plt.subplots(figsize=(6, 6))
    ax = plt.subplot(projection='polar')

    # Plot bars and points at the end to make them round
    for i, (block, percentage) in enumerate(percentages.items()):
        ax.barh(ys[i], percentage * 2 * pi, left=left, height=0.5, color=colors[i % len(colors)], label=block)
        ax.text(percentage + left + 0.02, ys[i], f'{percentage:.0%}', va='center', ha='left', color='black', fontsize=12)

    plt.ylim(-1, len(percentages))

    # Custom legend
    ax.legend(loc='center', frameon=True) 

    # Clear ticks and spines
    plt.xticks([])
    plt.yticks([])
    ax.spines.clear()
    plt.title(name, fontsize=15, pad=20, color="black")

    plt.show()
```

```python
#create a circular bar chart for accuracy by condition
from math import pi
plot_circular_bar_graph(test['correct'].groupby(test['condition']).mean(), 'Accuracy by Condition')
```
